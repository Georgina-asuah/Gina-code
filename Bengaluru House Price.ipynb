```python
def convert_sqt_to_num(x):
    token =x.split('-')
    if len(token) == 2:
        return (float(token[0]) + float(token[1])) /2
    try:
        return float(x)
    except:
        return None
    
```

```python
convert_sqt_to_num('213 stq. Meter')
```

```python
house2 =house1.copy()
```

```python
house2['total_sqft'] =house2['total_sqft'].apply(convert_sqt_to_num)
```
